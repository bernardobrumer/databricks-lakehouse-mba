01. Acessar o dataset acme_orders.csv

```python
df_ordens = spark.read \
    .format('csv') \
    .option('header', 'true') \
    .option('inferSchema', 'true') \
    .load('/Volumes/valeparaopeba/landing/vol_landing/mba/trabalho/acme_orders.csv')
```

```python
display(df_ordens.limit(10))
```

02. Acessar o dataset

```python
df_clientes = spark.read \
    .format('csv') \
    .option('header', 'true') \
    .option('inferSchema', 'true') \
    .load('/Volumes/valeparaopeba/landing/vol_landing/mba/trabalho/acme_customers.csv')
```

```python
display(df_clientes.limit(10))
```

03. Realizando a junção com o método .join()

```python
df_ordens = df_ordens.join(df_clientes, ['customer_id'], "left")
```

04. Aplicar pelo menos 1 filtro (usando WHERE) python: .filter() / .where()

```python
df_ordens = df_ordens.filter(df_ordens.customer_state == 'SP')
```

```python
df_ordens.select("customer_state").distinct().show()
```

```python
display(df_ordens)
```

05. Fazer pelo menos 1 agregação que mude a granularidade dos dados (usando GROUP BY) python: groupby().agg()

```python
from pyspark.sql.functions import col, sum, count
```

```python
df_agg = df_ordens \
    .groupBy("customer_city") \
    .agg(count("order_id").alias("total_pedidos")) \
    .sort(col("total_pedidos"), ascending=False)
```

```python
display(df_agg)
```

06. Acrescentar pelo menos 1 coluna calculada após a agregação

```python
# Calcula o total de pedidos (um DataFrame com uma linha)
total_pedidos = df_agg.agg(sum("total_pedidos").alias("total")).collect()[0]["total"]

# Cria a coluna percentual
df_agg = df_agg.withColumn("percentual", (col("total_pedidos") / total_pedidos) * 100)
```
